# Read in data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv', index_col=0)
df
```

```python
df.describe()
```

# Finding Mean Age

```python
print('The mean age is ' + str(df['Age'].mean()))
```

# Finding Oldest Player

```python
print("The oldest player is", df["Name"].iloc[df["Age"].idxmax()], "who is", df["Age"].max(), "years old.")
```

# Finding the highest salary

```python
best = -1
for salary in df['Release Clause'].dropna():
    temp = -1;
    if salary.endswith("K"):
        temp = float(salary.strip("€").strip("K"))*1000
    elif salary.endswith("M"):
        temp = float(salary.strip("€").strip("M"))*1000000
    if best < temp:
        best=temp
if best!= -1:                   
    print('The highest slary is €'+ str(best))
else:
    print('No data')
       
#pd.to_numeric(df["Release Clause"].str.strip("€").str.strip("M"), errors="coerce").max(skipna=True)*1000000
```

# Histogram

```python
"""
This histogram shows the age distrubutions of all the soccer players, we can see how some people became soccer players at just the 
age of 15 and some players all the way until 45 years of age. We can also see how most of the soccer players are between the ages
20 to 25.
"""
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ages=df['Age']
plt.hist(ages,bins=[15,20,25,30,35,40,45])
ax.set_title('Number of players per age')
ax.set_xlabel('Age')
ax.set_ylabel('Player Count')
plt.show()
```

# ScatterPlot

```python
'''
This ScatterPlot shows how that the more players generally tend to do short passes the better their overall performance increases,
the scatterplot's line of best fit is going linearly upwards diagonally showing there is a pretty strong correlation
'''
plt.scatter(x=df['ShortPassing'], y=df['Overall'])
plt.xlabel("ShortPassing")
plt.ylabel('Overall')
plt.title('ShortPassing vs Overall Performence')
plt.show()
```
